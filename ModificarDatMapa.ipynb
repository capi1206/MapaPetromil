```python
def dif_lin(p1,p2,pn):
    a,b,c,d,x,y=p1[0],p1[1],p2[0],p2[1],pn[0],pn[1]
    #print(a,b,c,d,x,y)
    top=(b*c) - (a*d) +(x*(d-b))+(y*(a-c))
    bot= ((b-d)**2) + ((c-a)**2)
    if bot==0:
        return 1
    return top/bot
def dis_pt(p1,p2,pn):
    a,b,c,d,x,y=p1[0],p1[1],p2[0],p2[1],pn[0],pn[1]
    return abs(a-x)+abs(c-x)+abs(b-y)+abs(d-y)
def mod_path2(longs, lats):
    if len(longs)<3:
        return longs, lats
    n_ind=[]
    i=0
    while i < len(longs)-2:
        diff=dif_lin((longs[i], lats[i]),(longs[i+2], lats[i+2]),(longs[i+1], lats[i+1]))
        dists=dis_pt((longs[i], lats[i]),(longs[i+2], lats[i+2]),(longs[i+1], lats[i+1]))
        if (diff <=0.16 and diff >= -0.16) or dists < 0.34:
            n_ind.append(i+1)
        i+=2    
    return [longs[k] for k in range(len(longs)) if k not in n_ind],[lats[k] for k in range(len(longs)) if k not in n_ind]

def mod_path(path, ind):
    if len(path)<=2 or ind>=len(path)  -2:
        return path
    else:
        a=path[ind]
        b=path[ind +2]
        x=path[ind+1]
        if dif_lin(a,b,x) <=0.05 and dif_lin(a,b,x) >= -0.05:
            n_path=path[:ind +1]+path[ind+2:]
            return mod_path(n_path, ind)
        else:
            return mod_path(path, ind+1)
```
